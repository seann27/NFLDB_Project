```python
from scrapers.GeneralScraper import Scraper
import re
```

```python
season = 2019
week = 9
```

```python
base_url = 'http://www.nfl.com/schedules/'
base_url += str(season)+'/reg'
base_url += str(week)
```

```python
weekly_games_soup = Scraper(base_url).get_soup()
```

```python
links = weekly_games_soup.findAll("div",{"data-gameid":re.compile(r'[0-9]{10}')})
```

```python
links
```

```python
gamelinks = []
```

```python
for link in links:
    gameid = link['data-gameid']
    gamelinks.append("http://www.nfl.com/liveupdate/game-center/"+str(gameid)+'/'+str(gameid)+"_gtd.json")
```

```python
gamelinks
```

```python
gameid = gamelinks[0]
```

```python
import urllib, json
with urllib.request.urlopen(gameid) as url:
    data = json.loads(url.read().decode())
```
