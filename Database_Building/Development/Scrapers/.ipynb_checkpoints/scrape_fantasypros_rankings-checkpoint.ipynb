```python
import pandas as pd
import numpy as np
import re
from scrapers.GeneralScraper import Scraper
from NFL_RefMaps import TableColumns,TeamDictionary
```

```python
flex = ['rb','wr','te','flex']
link = "https://www.fantasypros.com/nfl/rankings/half-point-ppr-rb.php"
page_soup = Scraper(link).page_soup
```

```python
metrics = [TableColumns().fantasy_pros['rankings']]
```

```python
data = page_soup.find('tbody')
players = data.findAll("tr",{"class":re.compile(r'mpb-player-')})
for idx,player in enumerate(players):
    info = player.findAll("a")
    pid = info[-1].get("class")[1]
    name = info[-1].get("fp-player-name")
    team = players[0].findAll("small")[0].text
    mets = np.array([pid,name,team])
    stats = player.findAll("td")
    opp = stats[3].text.strip().split()
    home_away = 1
    if len(opp) == 0:
        print("Error detected in ",link," -> player ",name," ... n/a provided for opponent")
        opp = 'n/a'
        home_away = 0
    elif opp[0] == 'at':
        home_away = 0
        opp = opp[1]
    else:
        opp = opp[1]
    mets = np.array([pid,name,team,opp,home_away])
    for i in range(4,len(stats)):
        mets = np.append(mets,stats[i].text)
    metrics.append(mets)
df = np.vstack(metrics)

pd.DataFrame(data=df[1:,1:],index=df[1:,0],columns=df[0,1:])
```
