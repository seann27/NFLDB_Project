```python
# returns quarter_end indicator, time left in half
def get_time_status(qtr,time):
    time = time.split(':')
    qtr = (int(qtr)%2)*15
    qtr_end = 0
    time_left = 0
    if(len(time)<2):
        qtr_end = 1
        time_left = 0
    else:
        seconds_remaining = float(int(time[1])/60)
        minutes_remaining = int(time[0])+qtr
        time_left = minutes_remaining+seconds_remaining
    return qtr_end,time_left

get_time_status('1','6:55')
```

```python
mpd = init_df()

mpd['game_id'] = get_gameid(nflapi)
mpd['home_team'],mpd['away_team'] = get_teams(nflapi)
mpd['date'] = get_date(nflapi)
drives = get_drive_ids(nflapi)
for drive in drives:
    plays = get_plays(drive,nflapi)
    for play in plays:
        # get play summary
        mpd['drive'].append(drive)
        mpd['play_id'].append(play)
        details = plays[play]
        mpd['sp'].append(details['sp'])
        mpd['qtr'].append(details['qtr'])
        mpd['down'].append(details['down'])
        mpd['time'].append(details['time'])
        qtr_end,time_left = get_time_status(details['qtr'],details['time'])
        mpd['quarter_end'].append(qtr_end)
        mpd['time_half'].append(time_left)
        mpd['ydstogo'].append(details['ydstogo'])
        mpd['ydsnet'].append(details['ydsnet'])
        mpd['desc'].append(details['desc'])
        mpd['posteam'].append(details['posteam'])
        mpd['yardline_100'].append(get_yardline_100(details['yrdln'],details['posteam']))
        
        # get qb action
        qb_kneel,qb_spike,qb_scramble,shotgun,no_huddle = get_qb_action(details['desc'])
        mpd['qb_kneel'].append(qb_kneel)
        mpd['qb_spike'].append(qb_spike)
        mpd['qb_scramble'].append(qb_scramble)
        mpd['shotgun'].append(shotgun)
        mpd['no_huddle'].append(no_huddle)
        
        # get penalty info
        penalty_yards, penalty_team, penalty_player_id, penalty_player_name = get_penalties(details['players'])
        mpd['penalty_yards'].append(penalty_yards)
        mpd['penalty_team'].append(penalty_team)
        mpd['penalty_player_id'].append(penalty_player_id)
        mpd['penalty_player_name'].append(penalty_player_name)
        mpd['yards_gained'].append(get_yards_gained(details['players'])+penalty_yards)
        
        # get rush and pass attempts
        rush_att,pass_att = get_rushpass_att(details['players'])
        mpd['rush_attempt'].append(rush_att)
        mpd['pass_attempt'].append(pass_att)
        
        # get sack info
        sack, sack_pid, sack_name = get_sack(details['players'])
        mpd['sack'].append(sack)
        mpd['sack_pid'].append(sack_pid)
        mpd['sack_name'].append(sack_name)
        
        # get touchdown info
        td, pass_td, rush_td, return_td = get_touchdowns(details['players'])
        mpd['touchdown'].append(td)
        mpd['pass_touchdown'].append(pass_td)
        mpd['rush_touchdown'].append(rush_td)
        mpd['return_touchdown'].append(return_td)
        
        # get special teams attempts
        xp_att, twopt_att, fg_att, kickoff_att, punt_att = get_special_teams_att(details['players'])
        mpd['extra_point_attempt'].append(xp_att)
        mpd['two_point_attempt'].append(twopt_att)
        mpd['field_goal_attempt'].append(fg_att)
        mpd['kickoff_attempt'].append(kickoff_att)
        mpd['punt_attempt'].append(punt_att)
        
        # get fumble
        mpd['fumble'].append(get_fumble(details['players']))
        
        # get pass result
        comp, incomp, passer_pid, passer_name, receiver_pid, receiver_name = get_pass_info(details['players'])
        mpd['complete_pass'].append(comp)
        mpd['incomplete_pass'].append(incomp)
        mpd['passer_player_id'].append(passer_pid)
        mpd['passer_player_name'].append(passer_name)
        mpd['receiver_player_id'].append(receiver_pid)
        mpd['receiver_player_name'].append(receiver_name)
        
mpd = pd.DataFrame.from_dict(mpd)
mpd
```
